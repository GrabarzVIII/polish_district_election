```python
import pandas as pd
```

```python
pres_df = pd.read_csv(r"C:\Users\kgrab\Desktop\Python Jupyte Notebook\polish_district_election\data_source\prez_2025_I.csv", 
                      sep=";")
pres_df.head()
```

```python
pres_df.info()
```

```python
pres_2025_data_set = pres_df[['TERYT Powiatu',
                              'Powiat',
                              'Liczba wyborców uprawnionych do głosowania',
                              'Liczba kart wyjętych z urny',
                              'NAWROCKI Karol Tadeusz',
                              'TRZASKOWSKI Rafał Kazimierz']]
pres_2025_data_set.head()
```

```python
pres_2025_data_set = pres_2025_data_set.rename(columns={'TERYT Powiatu':'TERYT',
                                                        'Powiat':'district',
                                                        'Liczba wyborców uprawnionych do głosowania':'entitled',
                                                        'Liczba kart wyjętych z urny':'voted'})
pres_2025_data_set['who_win'] = pres_2025_data_set.apply(
                                lambda x: 'PiS' if x['NAWROCKI Karol Tadeusz'] > x['TRZASKOWSKI Rafał Kazimierz'] else 'PO',
                                axis=1
                                                        )

pres_2025_data_set = pres_2025_data_set.drop(columns=['NAWROCKI Karol Tadeusz','TRZASKOWSKI Rafał Kazimierz'], axis=1)
pres_2025_data_set['voter_turnout%'] = round((pres_2025_data_set['voted']/pres_2025_data_set['entitled'])*100,2)
pres_2025_data_set['TERYT'] = pres_2025_data_set['TERYT'].astype('Int64').astype(str)

pres_2025_data_set.head()
```

```python
pres_2025_data_set.to_csv("pres_2025_I_data_set",sep=";")
```
